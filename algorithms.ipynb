```python
# SORTING ALGORITHMS

def merge_sort(arr):
    print(f"merge_sort({arr})")
    
    # base case
    if len(arr) <= 1:
        return arr
    
    # recursive case
    else:
        # split array
        mid = len(arr) // 2
        left_half = arr[:mid]
        right_half = arr[mid:]
        
        # call merge sort on each half
        left_half = merge_sort(left_half)
        right_half = merge_sort(right_half)
        
        sorted_arr = merge(left_half, right_half)
        
        return sorted_arr

def merge(left, right):
    print(f"merge({left}, {right})")
    output = []
    i, j = 0, 0
    
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            output.append(left[i]) 
            i += 1
        else:
            output.append(right[j])
            j += 1
    
    if i < len(left):
        output.extend(left[i:])
    
    if j < len(right):
        output.extend(right[j:])
        
    return output

def quick_sort(arr):
    print(f"quick_sort({arr})")
    if len(arr) <= 1:
        return arr

    pivot = arr[len(arr) // 2]
    smaller, equal, larger = [], [], []

    for element in arr:
        if element < pivot:
            smaller.append(element)
        elif element == pivot:
            equal.append(element)
        else:
            larger.append(element)

    return quick_sort(smaller) + equal + quick_sort(larger)

arr = [2, 1, 3, 5, 9, 7, 6, 4, 8, -1]
sorted_arr = quick_sort(arr)
print(arr)
print(sorted_arr)    
```

```python
# SEARCHING ALGORITHMS

def binary_search(k, arr):
    print(f'binary_search({k}, {arr})')
    
    # base case
    if len(arr) == 0:
        return False
    
    # recursive case
    mid = len(arr) // 2
    if k == arr[mid]:
        return True
    elif k <= arr[mid]:
        return binary_search(k, arr[:mid])
    else: # k > arr[mid]
        return binary_search(k, arr[mid+1:])
    
    
arr = list(range(10))
binary_search(-1, arr)
    
```
